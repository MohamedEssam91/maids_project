```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("train_maids.xlsx")
```

```python
# Summary statistics
print(df.describe())
```

```python
for i in df.columns:
    null_counts = df[i].isnull().sum()
    print(f"{i} column has ", null_counts, "nulls")
```

```python
# Get column names
features = df.columns.drop('price_range')
```

```python
# Bar graph to know how many ranges in the data and quanitify it
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='price_range')
plt.title('Distribution of Price Range')
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.show()
```

```python
# Correlation matrix
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatter plot analysis for each column but focus on battery_power, px_height, px_width, ram since they are the highest correlated to price_range
for feature in features:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x=feature, y='price_range')
    plt.title(f'{feature.capitalize()} vs Price Range')
    plt.xlabel(feature.capitalize())
    plt.ylabel('Price Range')
    plt.show()
```
